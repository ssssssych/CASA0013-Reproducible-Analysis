#### House price annual change

```python
#Folding Line Chart
label_name = ['2016','2017','2018','2019','2020','2021','2022']

num_list=[506566,522732,517786,	507910,	521139, 537759,	563879]

line = plt.plot(label_name,num_list)

plt.ylabel("Price")

plt.xlabel("Year")

plt.title('london housing prices annual change')

plt.legend()

plt.show()
```

We can see that house prices in London are on an upward trend and have continued to rise at a high rate over the last three years. In 2019, there was a brief reduction in house prices, presumably due to special reasons such as the epidemic

```python
data['2016'].describe()
```

#### 2016 london house price

```python
# 
fig, ax = plt.subplots()

data['2016'].hist(color='#1CD0EF')

ax.set_xlabel('house price rank')

ax.set_ylabel('Frequency')

ax.set_title('2016 london house price distribution')

plt.show()
```

#### 2022 london house price

```python
data['2022'].describe()
```

```python
# 
fig, ax = plt.subplots()

data['2022'].hist(color='#1CD0EF')

ax.set_xlabel('house price rank')

ax.set_ylabel('Frequency')

ax.set_title('2022 london house price distribution')

plt.show()
```

We can see that comparing the bar chart for 2016 and 2022 shows that the proportion of houses with an average house price of around 500thousand has increased, while the proportion of higher priced houses has decreased.

#### 2016-2022 london house price Box plot

```python
box2016,box2017,box2018,box2019,box2020,box2021,box2022 = data['2016'],data['2017'],data['2018'],data['2019'],data['2020'],data['2021'],data['2022']

plt.figure(figsize = (10,5))

plt.title('2016-2022 London prices')
lable = '2016','2017','2018','2019','2020','2021','2022'
plt.boxplot([box2016,box2017,box2018,box2019,box2020,box2021,box2022],labels=lable)
plt.show()
```

```python
The outliers in the graph are the higher values relative to the average house price. And as can be seen from the box plot for 2016 to 2022, there is a trend towards a normal distribution of house prices in London and the higher house prices are reflected.
```
